```python
print(f'Init')
import cv2
import matplotlib.pyplot as plt
pth = 'static/images/best_home.png'
img = cv2.imread(pth)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
def change_px(px):
    if(px[0] == 73 and px[1] == 91 and px[2] == 109):
        return False
    elif(px[0] == 135 and px[1] == 205 and px[2] == 188):
        return False
    else:
        return True


for i in range(2040):
    for j in range(2210):
        px = img[i][j]
        if(change_px(px) == True):
            img[i][j] = [0,0,0]
            break
        
```
